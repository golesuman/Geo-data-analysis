```python
import osmnx as ox
import networkx as nx
```

```python
ox.config(log_console=True, use_cache=True)
place = 'Lalitpur, Nepal'
mode = 'walk'
optimizer = 'time'
tags = {'amenity': 'hospital',
       'category':'general'}
```

```python
hospital = ox.geometries_from_address(place, tags = tags )
hospital.head()
```
